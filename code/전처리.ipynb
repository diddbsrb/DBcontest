```python
import pandas as pd
import numpy as np
```

```python
cancer_df = pd.read_csv("전체암발생률.csv", encoding='cp949')
```

```python
cancer_df
```

```python
population_df = pd.read_csv("전체인구.csv", encoding='cp949')
```

```python
population_df
```

```python
# 첫 번째 행 제거 (설명 행일 가능성 있음)
cancer_df = cancer_df.iloc[1:].reset_index(drop=True)

# 컬럼명 변경
cancer_df.rename(columns={"연령별": "AgeGroup", "성별": "Sex"}, inplace=True)

# 데이터를 long format으로 변환
cancer_melted = cancer_df.melt(id_vars=["Sex", "AgeGroup"], var_name="Year", value_name="Value")
```

```python
cancer_melted
```

```python
cancer_melted = cancer_melted.loc[34:,:]
```

```python
cancer_melted
```

```python
# 연도 값 추출 (연도만 숫자로 변환)
cancer_melted["Year"] = cancer_melted["Year"].astype(str).str.extract(r'(\d{4})').astype(int)
# 발생자 수(`CancerCase`)와 조발생률(`IncidenceRate`)을 구분하는 `Metric` 컬럼 생성
cancer_melted["Metric"] = cancer_melted["Value"].apply(lambda x: "CancerCase" if str(x).replace('.', '').isdigit() and '.' not in str(x) else "IncidenceRate")
cancer_melted["Value"] = pd.to_numeric(cancer_melted["Value"])
```

```python
# 발생자 수와 조발생률을 각각 컬럼으로 변환
cancer_final = cancer_melted.pivot_table(index=["Sex", "AgeGroup", "Year"], columns="Metric", values="Value").reset_index()
```

```python
# 성별 변환 ("남자" → "M", "여자" → "F")
cancer_final["Sex"] = cancer_final["Sex"].replace({"남자": "M", "여자": "F"})

# 연령대 표기 변경 ("20-24세" → "20~24세", "25-29세" → "25~29세")
cancer_final["AgeGroup"] = cancer_final["AgeGroup"].str.replace("-", "~")
```

```python
cancer_final
```

```python
population_df
```

```python
# 컬럼명 변경
population_df.rename(columns={"연령별": "AgeGroup", "항목": "Sex_Info"}, inplace=True)

# 성별 구분 ("총인구_남자(명)" → "M", "총인구_여자(명)" → "F")
population_df["Sex"] = population_df["Sex_Info"].apply(lambda x: "M" if "남자" in x else "F")
```

```python
population_df
```

```python
# 필요 없는 컬럼 제거
population_df.drop(columns=["행정구역별(읍면동)", "Sex_Info"], inplace=True)

# 데이터를 long format으로 변환 (연도별 인구 데이터 펼치기)
population_melted = population_df.melt(id_vars=["Sex", "AgeGroup"], var_name="Year", value_name="Population")
```

```python
population_melted
```

```python
# 연도 값 정리
population_melted["Year"] = population_melted["Year"].astype(str).str.extract(r'(\d{4})').astype(int)

# Population 컬럼 숫자로 변환
population_melted["Population"] = pd.to_numeric(population_melted["Population"], errors="coerce")
```

```python
population_melted
```

```python
cancer_final
```

```python
### 3️⃣ 암 발생 데이터와 인구 데이터 병합 ###
df_final = pd.merge(cancer_final, population_melted, on=["Sex", "AgeGroup", "Year"], how="inner")

# `Group` 컬럼 생성 (예: M20_24, F25_29)
df_final["Group"] = df_final["Sex"] + df_final["AgeGroup"].str[:-1]

# 최종 컬럼 정리
df_final = df_final[["Year", "Group", "IncidenceRate", "Population", "CancerCase"]]
```

```python
df_final['Group'] = df_final['Group'].str.replace('~', '_')
```
